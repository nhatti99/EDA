```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.feature_selection import f_classif, f_regression
import statsmodels.api as sm
```

```python
df = pd.read_csv("WineQT.csv")
```

```python
# Generate raw data report
def rawdata_report(df, rawdata_report_file):
    with open(rawdata_report_file, 'w') as t:
        t.write("Raw Data Report\n")
        t.write("-----------------\n\n")
        
        #Data Shape
        t.write("Data Shape:\n")
        t.write(f"{df.shape}\n\n")
        
        # Null values
        t.write("Missing values:\n")
        t.write(f"{df.isnull().sum()}\n\n")
        
        # Basic Statistics
        t.write("Data Info:\n")
        #buffering (redirecting output)
        df.info(buf=t) 
        t.write("\nStatistics Summary:\n")
        t.write(f"{df.describe(include='all')}\n\n")
```

```python
# Clean the data 
def data_clean(df):
    df.drop_duplicates(inplace = True)
    
    #Remove serial numbers and column with all unique values
    for col in df.columns:
        if df[col].nunique() == len(df):
            df.drop(col, axis=1, inplace=True)
            
    return df
```

```python
#Preprocessing the data
def data_preprocess(df):
    #Null value imputation
    df.fillna(df.mean(), inplace=True)  #If the variable is categorical but represent numerically, 
                                        #the mean will be calculatedfor that variable and any missing 
                                        #values will be filled with the mean value
    df.fillna(df.mode().iloc[0],inplace=True)
    
    categorical_col = df.select_dtypes(include=['object']).columns
    df = pd.get_dummies(df, columns = categorical_col, drop_first=True)
    
    #Scaling
    scaler = StandardScaler()
    numeric_col = df.select_dtypes(include=[np.number]).columns
    df[numeric_col] = scaler.fit_transform(df[numeric_col])
    
    return df
```

```python
def data_explore(df, output_column, rawdata_report_file):
    with open(rawdata_report_file, 'w') as t:
        t.write("Data Preprocessed Report\n")
        t.write("-----------------------------\n\n")
        t.write("Data Shape:\n")
        t.write(f"{df.shape}\n\n")
        t.write("Missing values:\n")
        t.write(f"{df.isnull().sum()}\n\n")
        t.write("Data Info:\n")
        df.info(buf=t)
        t.write("\nStatistics summary:\n")
        t.write(f"{df.describe()}\n\n")
        corr_matrix = df.corr()
        t.write("Correlation matrix:\n")
        t.write(f"{corr_matrix}\n\n")
        
        #Statistical tests
        t.write("Statistical Tests:\n")
        X = df.drop(output_column, axis = 1)
        y = df[output_column]
        
        #F-test for classification data
        if df[output_column].dtype == 'object':
            f_values, p_values = f_classif(X,y)
            # F_test for regression data
        else:
            f_values, p_values = f_regression(X,y)
            
        stat_test_result = pd.DataFrame({"Feature":X.columns, "F_Value":f_values, "P_Value":p_values})
        t.write(f"{stat_test_result}\n\n")
        
    return corr_matrix
```

```python
# Data Vizualization
def data_visualize(df, corr_matrix):
    # For numerical features
    sns.pairplot(df)
    plt.show()
    
    #Heatmap
    plt.figure(figsize=(12,8))
    sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
    plt.show()
    
    numeric_cols = df.select_dtypes(include=[np.number]).columns
    for col in numeric_cols:
        plt.figure()
        sns.histplot(df[col],kde=True)
        plt.title(f"Distribution of {col}")
        plt.show()
```

```python
# creating pipeline
def data_pipeline(df, output_column, rawdata_report_file="raw_data_report.txt", preprocessed_report_file="preprocessed_data_report.txt"):
    print("Original Data Shape:", df.shape)
    
    rawdata_report(df, rawdata_report_file)
    print(f"Raw data report file saved to {rawdata_report_file}")
    
    df_cleaned = data_clean(df)
    print("Cleaned Data Shape", df_cleaned.shape)
    
    df_preprocessed = data_preprocess(df_cleaned)
    print("Preprocessed Data Shape:",df_preprocessed.shape)
    
    corr_matrix = data_explore(df_preprocessed, output_column, preprocessed_report_file)
    data_visualize(df_preprocessed, corr_matrix)
    print(f"Processed data report saved to {preprocessed_report_file}")
```

```python
data_pipeline(df, output_column='quality')
```
